```python
import cv2
from pathlib import Path
from ultralytics import YOLO

# Correct the file path using raw string literal
model_path = r'C:\MY\Miniconda_py\Drowsiness\YOLO-Drowsiness-Detection\runs\detect\train6\weights\best.pt'  # Use raw string to handle backslashes

# Ensure that the model path is correct
print(f"Loading model from: {model_path}")

try:
    # Load your YOLOv8 model
    model = YOLO(model_path)
    print("Model loaded successfully.")
except Exception as e:
    print(f"Error loading model: {e}")

def predict_image(image_path, model):
    # Load the image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Could not read image {image_path}")
        return

    # Perform inference
    results = model(img)

    # Since results is a list, process each result
    for result in results:
        # Print detailed results
        print("Results:")
        print(f"Image Path: {result.path}")
        print(f"Original Shape: {result.orig_shape}")
        print(f"Detected Boxes: {result.boxes}")
        print(f"Speed: {result.speed}")

        # Extract boxes and their attributes
        if result.boxes:
            boxes = result.boxes.xyxy.cpu().numpy()
            classes = result.boxes.cls.cpu().numpy()
            confidences = result.boxes.conf.cpu().numpy()
            
            # Print each detection
            for box, cls, conf in zip(boxes, classes, confidences):
                print(f"Box: {box}, Class: {cls}, Confidence: {conf}")

        else:
            print("No detections were made.")

        # Annotate the image with bounding boxes
        annotated_img = result.plot()

        # Display the image with bounding boxes
        cv2.imshow('Prediction', annotated_img)
        cv2.waitKey(0)

        # Save the annotated image
        output_path = f"{Path(image_path).stem}_pred.jpg"
        cv2.imwrite(output_path, annotated_img)
        print(f"Saved annotated image to {output_path}")

# Define the directory containing your images
image_dir = r'C:\MY\Miniconda_py\Drowsiness\YOLO-Drowsiness-Detection'  # Use raw string to handle backslashes

# Ensure that the image directory path is correct
print(f"Loading images from: {image_dir}")

# Iterate over images in the directory and perform predictions
try:
    for image_path in Path(image_dir).glob('*.jpg'):  # Adjust the extension as needed
        print(f"Processing image: {image_path}")
        predict_image(str(image_path), model)
except Exception as e:
    print(f"Error processing images: {e}")

# Close all OpenCV windows
cv2.destroyAllWindows()

```

```python
import cv2
from ultralytics import YOLO

# Correct the file path using raw string literal
model_path = r'C:\MY\Miniconda_py\Drowsiness\YOLO-Drowsiness-Detection\runs\detect\train6\weights\best.pt'

# Ensure that the model path is correct
print(f"Loading model from: {model_path}")

try:
    model = YOLO(model_path)
    print("Model loaded successfully.")
except Exception as e:
    print(f"Error loading model: {e}")

def predict_frame(frame, model):
    results = model(frame)

    for result in results:
        print("Processing result...")
        if result.boxes:
            boxes = result.boxes.xyxy.cpu().numpy()
            classes = result.boxes.cls.cpu().numpy()
            confidences = result.boxes.conf.cpu().numpy()

            for box, cls, conf in zip(boxes, classes, confidences):
                print(f"Box: {box}, Class: {cls}, Confidence: {conf}")

        annotated_frame = result.plot()
        return annotated_frame

def main():
    cap = cv2.VideoCapture(0)

    if not cap.isOpened():
        print("Error: Could not open webcam.")
        return

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error: Could not read frame.")
            break

        print(f"Frame shape: {frame.shape}, Frame type: {type(frame)}")

        try:
            annotated_frame = predict_frame(frame, model)
            cv2.imshow('Live Prediction', annotated_frame)

        except Exception as e:
            print(f"Error during prediction or displaying frame: {e}")
            continue

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

main()
```
